```python
import yaml

merge_yaml = """
---
configuration:
  output_directory: merged
  checkpoint: false
  property_types:
    # # define the type for non-canonical node/edge properties
    # combined_score: 'xsd:float'
    # confidence_score: 'xsd:float'
    # neighborhood: 'xsd:float'
    # neighborhood_transferred: 'xsd:float'
    # fusion: 'xsd:float'
    # cooccurence: 'xsd:float'
    # homology: 'xsd:float'
    # coexpression: 'xsd:float'
    # coexpression_transferred: 'xsd:float'
    # experiments: 'xsd:float'
    # experiments_transferred: 'xsd:float'
    # database: 'xsd:float'
    # database_transferred: 'xsd:float'
    # textmining: 'xsd:float'
    # textmining_transferred: 'xsd:float'

merged_graph:
  name: Elenas awesome KG
  source:
    hp:
      name: "HPO"
      input:
        format: tsv
        filename:
          - hp_kgx_tsv_nodes.tsv
          - hp_kgx_tsv_edges.tsv
    mondo:
      name: "Mondo"
      input:
        format: tsv
        filename:
          - mondo_kgx_tsv_nodes.tsv
          - mondo_kgx_tsv_edges.tsv
    GO:
      name: "GO"
      input:
        format: tsv
        filename:
          - go_kgx_tsv_nodes.tsv
          - go_kgx_tsv_edges.tsv
    CHEBI:
      name: "CHEBI"
      input:
        format: tsv
        filename:
          - chebi_kgx_tsv_nodes.tsv
          - chebi_kgx_tsv_edges.tsv

  operations:
    - name: kgx.graph_operations.summarize_graph.generate_graph_stats
      args:
        graph_name: Elenas nobel prize winning KG
        filename: merged_graph_stats.yaml
        node_facet_properties:
          - provided_by
        edge_facet_properties:
          - provided_by
          - source
  destination:
    merged-kg-tsv:
      format: tsv
      filename: merged-kg
"""

parsed_yaml_content = yaml.safe_load(merge_yaml)

# Define the local file path to save the YAML content
local_file_path = "merge.yaml"

# Save the YAML content to the local file system
with open(local_file_path, "w") as file:
    yaml.dump(parsed_yaml_content, file, default_flow_style=False)

print(f"YAML content saved to {local_file_path}")
```
